```python
import pandas as pd 
import numpy as np 
```

```python
mov = pd.read_csv("data/tmdb_5000_movies.csv")
```

```python
cred = pd.read_csv("data/tmdb_5000_credits.csv")
```

```python
#merging the datasets
mov = mov.merge(cred,on="title")
mov.columns
```

```python
mov.isnull().sum() 
```

```python
mov = mov[['movie_id','title','genres','cast','crew','keywords','vote_average']] #only essential columns
mov
```

```python
#converting stringified json into usable values
import ast #Abstract Syntax Trees
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

#mov['genres'] = mov['genres'].apply(convert)
mov['cast'] = mov['cast'].apply(convert)
mov['crew'] = mov['crew'].apply(convert)
mov
```

```python
def convert_cast(text):
    return text[:3]
mov['cast'] = mov['cast'].apply(convert_cast)
mov
    
```
