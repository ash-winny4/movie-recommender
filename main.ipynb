```python
import pandas as pd 
import numpy as np 
```

```python
mov = pd.read_csv("data/tmdb_5000_movies.csv")
```

```python
cred = pd.read_csv("data/tmdb_5000_credits.csv")
```

```python
#merging the datasets
mov = mov.merge(cred,on="title")
mov.columns
```

```python
mov.isnull().sum() 
```

```python
mov = mov[['movie_id','title','genres','cast','crew','keywords','vote_average','overview']] #only essential columns
mov
```

```python
#converting stringified json into usable values
import ast #Abstract Syntax Trees
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

mov['genres'] = mov['genres'].apply(convert)
mov['cast'] = mov['cast'].apply(convert)
mov['crew'] = mov['crew'].apply(convert)
mov
```

```python
def convert_cast(text):
    return text[:3]
mov['cast'] = mov['cast'].apply(convert_cast)
mov
    
```

```python
#mov['genres'] = mov['genres'].apply(convert)
mov['keywords'] = mov['keywords'].apply(convert)
mov
```

```python
# addin tags 
mov['tags'] = mov['genres'] + mov['cast'] + mov['crew'] + mov['keywords']
mov
```

```python
#mov['tags'] = mov['tags'].apply(lambda x: " ".join(x)) #list to string cuz ML Models use string matching
#mov_def = mov[['movie_id','title','tags']].copy()
#mov_def['tags'] = mov_def['tags'].apply(lambda x: x.lower())
print(type(mov['tags'][0]))
mov['tags'] = mov['tags'].apply(lambda x: " ".join(x))
mov['tags'] = mov['tags'].apply(lambda x: x.lower())
#mov_def
```

```python
mov_def = mov[['movie_id','title','tags']].copy()
mov_def
```

```python
#count vectorizer to convert string data into numbers (each movie will be a point in vector space) *imp*
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(mov_def['tags']).toarray()
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
import pickle

pickle.dump(mov, open('movies.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
```
